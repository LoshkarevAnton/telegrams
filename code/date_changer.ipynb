# Подготовка средств замены дат: функции и словаря

```python
months = {}
mnth = ['янв', 'фев','мар', 'апр', 'май', 'июн', 'июл', 'авг', 'сен', 'окт', 'ноя', 'дек']
i = 0
for m in mnth:
  i += 1
  months[m] = "{:02d}".format(i)
months['мая'] = "05"
months
```

```python
def datechanger(date):
  delimtr = ''
  date = str(date)
  s = re.match('[0-9xх][0-9xх]. ?[0-9xх][0-9xх]. ?[0-9xх][0-9xх][0-9xх][0-9xх]', date)
  if s != None:
    #print(date)
    date = ''.join(s.group(0).split())
  else:
    #print(date)
    delimtr = ''
    day, month, year = 'xx', 'xx', 'xxxx'

    y = re.findall('[0-9][0-9][0-9][0-9]', date)
    if len(y) == 1:
      year = y[0]
    if len(y) == 2:
      year = y[0][:-1]+'x'

    m = re.findall('[а-яА-Я]+', date)
    for en in m:
      for mon in months:
        if mon in en:
          delimtr = en
          month = months[mon]

    if delimtr != '':
      da_t = date.split(delimtr)
      for elem in da_t:
        d = re.findall('[0-9][0-9]?', elem)
        if len(d) == 1:
          day = d[0]

    date = day + '.' + month + '.' + year
  return date
```

## Применение средств замены дат и сохранение полученных результатов в другую таблицу

```python
df['Дата_расчищено'] = df['date'].apply(datechanger)
df
```
